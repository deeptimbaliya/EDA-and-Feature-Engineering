## Feature Engineering and data visulitation 


```python
df["average_score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
df.head()
```

```python
df[["math score","reading score","writing score"]].mean()
```

```python
sns.boxplot(x="gender", y="average_score", data=df)
plt.title("Gender vs Average Score")
plt.show()
```

### data shows thats female is good score then male 


```python
sns.boxplot(x="test preparation course", y="average_score", data=df)
plt.title("Impact of Test Preparation on Performance")
plt.show()
```

### Data shows that the student whi completed the test prepartion course is get good marks 

```python
plt.figure(figsize=(10,5))
sns.barplot(x="parental level of education", y="average_score", data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs Student Performance")
plt.show()
```

### data shows that students whose parents is more educated that get good marks 

```python
plt.figure(figsize=(10,5))
sns.barplot(x="race/ethnicity", y="average_score", data=df)
plt.xticks(rotation=45)
plt.title("Race/Ethnicity vs Student Performance")
plt.show()
```

```python
sns.heatmap(df[["math score","reading score","writing score"]].corr(), 
            annot=True, cmap="coolwarm")
plt.title("Correlation Between Subjects")
plt.show()
```
